Introducing Marital Status 


```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
df['Marital'] = df.apply(get_marital_status, axis=1)
```

```python
#Plotting MntTotal vs Marital Status to decipher which group of people are more inclined towards buying goods
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital', y='MntTotal', data=df, palette='viridis')
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
```

There are 3 features that reflect if a person is single (Single, Divorced, Widow) and 2 features if a person is in relationship (Together, Married). We will add an additional feature 'In_relationship'. This feature will equal 1 if a customer's marital status is 'Married' or 'Together' and 0 in all other cases.



```python
def get_relationship(row):
    if row['marital_Married'] == 1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0 
df['In_relationship'] = df.apply(get_relationship, axis=1)
df.head() 
```

K-Means Clustering : Machine Learning Algorithm to group similar data into a cluster, works well for large datasets


Standardsiing Data

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship']
df_scaled = df.copy()
df_scaled[cols_for_clustering] = scaler.fit_transform(df[cols_for_clustering])
df_scaled[cols_for_clustering].describe()
```

Mean Values ~ 0. Standard Deviation close to 1.
Data replaced by z-scores.

The elbow method is a technique used to determine the optimal number of clusters (K) for K-means clustering algorithm.
We get this by observing the graph plot to see where the variance drastically drops and stays near consistent throughout rest of plot.


```python
X = df_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
plt.figure(figsize=(6,4))
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

K should be 4 or 5.

Let us try Silhouette Score Analysis


```python
from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters = K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,5])
plt.plot(range(2,10), silhouette_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

So we can assume that K(No.of Clusters) should be set to 4.


```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(df_scaled[cols_for_clustering])
df_scaled['Cluster'] = model.predict(df_scaled[cols_for_clustering])
```

Time for Visuals of Clusters

```python
df['Cluster'] = df_scaled.Cluster
df.groupby('Cluster')[cols_for_clustering].mean()
```
